```python
Cell 1: Basic Structure of while Loop     
```

```python
# while loop runs as long as the condition is True

count = 1

while count <= 5:
    print("Count:", count)
    count += 1          # Important: Update the condition variable

print("Loop finished!")
```

```python
Cell 2: Using while Loop with Counter
```

```python
# Counting and accumulating using while loop

total = 0
i = 1

while i <= 5:
    total += i
    print(f"i = {i}, Running Total = {total}")
    i += 1

print("\nFinal Total:", total)
```

```python
Cell 3: Processing Until a Condition is Met
```

```python
# Keep processing until average reaches a target

marks = [45, 78, 92, 65, 88, 55]
total = 0
count = 0
index = 0
target_average = 70

while index < len(marks):
    total += marks[index]
    count += 1
    average = total / count
    
    print(f"After adding mark {marks[index]} → Average = {average:.2f}")
    
    if average >= target_average:
        print("Target average reached!")
        break
    
    index += 1
```

```python
Cell 4: Using break in while Loop
```

```python
# Stop the loop when a specific condition is found

numbers = [10, 25, 30, 45, 50, 65]
index = 0

while index < len(numbers):
    if numbers[index] > 50:
        print(f"First number greater than 50 found: {numbers[index]}")
        break
    index += 1
```

Cell 5: Using continue in while Loop

```python
# Skip certain values using continue

marks = [85, 45, 92, 30, 78, 55]
index = 0

while index < len(marks):
    if marks[index] < 50:
        index += 1
        continue                    # Skip this iteration
    
    print(f"Passed mark: {marks[index]}")
    index += 1
```

```python
Cell 7: for vs while – When to Use What
```

```python
marks = [85, 90, 78, 92, 88]

print("=== Using for loop (Recommended for lists) ===")
total = 0
for mark in marks:
    total += mark
print("Total:", total)

print("\n=== Using while loop ===")
total = 0
i = 0
while i < len(marks):
    total += marks[i]
    i += 1
print("Total:", total)
```

```python
Common Mistakes
```

```python
print("=== Common Mistakes with While Loops ===\n")

# ========== Mistake 1: Infinite Loop (Most Dangerous) ==========
count = 1

# while count <= 5:          # ❌ Missing update → Infinite loop!
#     print(count)

count = 1
while count <= 5:
    print("Count:", count)
    count += 1               # ✅ Must update the condition variable


# ========== Mistake 2: Using = instead of == ==========
marks = [85, 90, 78]
index = 0

# while index = 0:           # ❌ SyntaxError
while index == 0:            # ✅ Use == for comparison
    print("First mark:", marks[index])
    index += 1


# ========== Mistake 3: Off-by-one error ==========
numbers = [10, 20, 30, 40, 50]
i = 0

# while i <= len(numbers):     # ❌ IndexError on last iteration
while i < len(numbers):
    print(numbers[i])
    i += 1


# ========== Mistake 4: Forgetting to update index ==========
data = [5, 10, 15]
i = 0

# while i < len(data):
#     print(data[i])
#     # Missing i += 1 → Infinite loop!

while i < len(data):
    print(data[i])
    i += 1                   # ✅ Must increment index


# ========== Mistake 5: Using while when for is simpler ==========
marks = [85, 90, 78, 92, 88]

# Unnecessarily complicated with while
i = 0
total = 0
while i < len(marks):
    total += marks[i]
    i += 1
print("Total using while:", total)

# Much cleaner with for loop
total = 0
for mark in marks:
    total += mark
print("Total using for  :", total)


# ========== Mistake 6: Not handling empty list ==========
marks = []

i = 0
while i < len(marks):
    print(marks[i])          # This won't run, which is fine
    i += 1

print("Mistake 6 handled - Empty list is safe with while")


print("\n=== All common while loop mistakes covered ===")
```

```python
# =====================================================
# MODULE 3 PART 2 - WHILE LOOP CODE TRACING QUIZ
# Predict the output of each question before running
# =====================================================

print("========== QUIZ START ==========\n")

# Q1: Basic while loop with counter
count = 1
print("Q1:")
while count <= 4:
    print(count)
    count += 1
print()


# Q2: Accumulation using while loop
numbers = [10, 20, 30, 40]
total = 0
i = 0
while i < len(numbers):
    total += numbers[i]
    i += 1
print("Q2 - Total:", total)
print()


# Q3: Using break in while loop
marks = [45, 78, 32, 90, 55]
i = 0
print("Q3:")
while i < len(marks):
    if marks[i] > 80:
        print("First mark > 80:", marks[i])
        break
    i += 1
print()


# Q4: Using continue in while loop
scores = [85, 45, 92, 30, 78]
i = 0
print("Q4:")
while i < len(scores):
    if scores[i] < 50:
        i += 1
        continue
    print("Passed:", scores[i])
    i += 1
print()


# Q5: Finding maximum using while loop
data = [78, 92, 85, 88, 75]
i = 0
highest = 0
while i < len(data):
    if data[i] > highest:
        highest = data[i]
    i += 1
print("Q5 - Highest:", highest)
print()


# Q6: Counting with condition
marks = [45, 78, 32, 90, 55, 88]
i = 0
pass_count = 0
while i < len(marks):
    if marks[i] >= 50:
        pass_count += 1
    i += 1
print("Q6 - Passed count:", pass_count)
print()


# Q7: Combined logic with while loop
students = [
    {"name": "Aarav", "scores": [85, 90, 78]},
    {"name": "Priya", "scores": [45, 55, 38]}
]
i = 0
print("Q7:")
while i < len(students):
    avg = sum(students[i]["scores"]) / len(students[i]["scores"])
    print(students[i]["name"], "-> Average:", round(avg, 2))
    i += 1
print()


print("========== QUIZ END ==========")
```
