```python
#Create a dictionary to store a student's name and marks entered by the user
student = {}
name = input("enter student name: ")
marks = int(input("enter marks: "))
student["Name"] = name
student["Marks"] = marks
print(student)
```

```python
#Dictionary of 5 fruits and their prices
fruits = {
    "Apple": 120,
    "Banana": 40,
    "Orange": 80,
    "Mango": 150,
    "Grapes": 90
}
print(fruits)
```

```python
#Display the salary of a specified employee
employees = {
    "R": 50000,
    "P": 60000,
    "A": 55000
}
name = input("Enter employee name: ")
if name in employees:
    print("Salary:", employees[name])
else:
    print("Employee not found")
```

```python
#Print all countries and capitals
countries = {
    "India": "New Delhi",
    "USA": "Washington DC",
    "Japan": "Tokyo",
    "France": "Paris"
}
for country, capital in countries.items():
    print(country, ":", capital)
```

```python
#Update the marks of a student
students = {
    "Rahul": 75,
    "Anita": 82
}
students["Rahul"] = 90
print(students)
```

```python
#Increase product quantity in inventory
inventory = {
    "Pen": 50,
    "Book": 20
}
product = "Pen"
increase = 10
inventory[product] += increase
print(inventory)
```

```python
#Delete a specific key if it exists
student = {
    "Name": "Riya",
    "Marks": 88,
    "Grade": "A"
}
key = input("Enter key to delete: ")

if key in student:
    del student[key]
    print("Key deleted")
else:
    print("Key not found")
print(student)
```
