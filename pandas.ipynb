```python
import pandas as pd 
raw_data = [  
    ['Alice', 'Engineering', 'Senior', 110000, 15000],
    ['Bob', 'Marketing', 'Junior', 70000, 5000],
    ['Charlie', 'Engineering', 'Senior', 100000, 12000],
    ['David', 'Marketing', 'Senior', 95000, 8000],
    ['Eva', 'Engineering', 'Junior', 85000, 6000],
    ['Frank', 'Marketing', 'Junior', 65000, 4000]
] 
df2 = pd.DataFrame(raw_data)
```

```python
df2
```

```python
df2.columns = ['Employee', 'Team', 'Level', 'Salary', 'Bonus']
```

```python
df2
```

```python
columns = ["Employee", "Team", "Level", "Salary", "Bonus"] 
df2 = pd.DataFrame(raw_data, columns=columns)
```

```python
df2
```

```python
df2['Team_Avg_Salary'] = df2.groupby('Team')['Salary'].transform('mean')
```

```python
print(df2)
```

```python
summary_diff_math = df2.groupby('Team', as_index=False).agg({ 
    'Salary': 'mean', 
    'Bonus': 'sum'    
}) 
print(summary_diff_math)
```

```python
import matplotlib.pyplot as plt
df_dept = df2.groupby('Team')['Salary'].mean()
df_dept.plot(kind="bar", color=["skyblue", "green"])
plt.title("Average Salary by Department")
plt.ylabel("Salary (rupee)")
plt.show()
```

```python
df2["Total_Pay"] = df2["Salary"] + df2["Bonus"] 
df2.plot(x="Employee", y="Total_Pay", kind="line", marker="*", color="green")
plt.title("Total Pay per Employee")
plt.ylabel("Total Pay (Rupee)")
plt.show()
```

```python
df2.plot(
    kind="scatter", x="Bonus", y="Salary", color="darkblue", s=100, marker="o")
plt.title("Salary vs. Bonus Correlation") 
plt.grid(True, linestyle="--", alpha=0.8)
plt.show()
```
