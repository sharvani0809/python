```python
# Looping through a list of marks

marks = [85, 90, 78, 92, 88, 75]

print("All marks:")
for mark in marks:
    print(mark)

print("\nDone looping!")
```

```python
# Using range() to loop a specific number of times

print("Numbers from 0 to 5:")
for i in range(6):
    print(i)

print("\nEven numbers from 0 to 10:")
for i in range(0, 11, 2):
    print(i)
```

```python
subjects = ["Math", "Physics", "Chemistry", "CS", "English"]
marks = [85, 90, 78, 92, 88]

print("Subject-wise marks:")
for index, subject in enumerate(subjects):
    print(f"{index + 1}. {subject}: {marks[index]}")
```

```python
print("=== Common Mistakes with For Loops ===\n")

# ========== Mistake 1: Modifying list while iterating ==========
marks = [85, 45, 78, 32, 92, 55]

# Wrong way (can skip elements or cause unexpected behavior)
# for m in marks:
#     if m < 50:
#         marks.remove(m)           # ❌ Dangerous!

# Correct way: Create a new list
passed_marks = []
for m in marks:
    if m >= 50:
        passed_marks.append(m)

print("Mistake 1 fixed - Passed marks:", passed_marks)


# ========== Mistake 2: Forgetting to initialize variables ==========
scores = [85, 90, 78, 92, 88]

# total = 0          # Must initialize before loop
total = 0
for score in scores:
    total += score

print("Mistake 2 fixed - Total:", total)


# ========== Mistake 3: Wrong use of range(len()) ==========
names = ["Aarav", "Priya", "Rahul"]

# print(names[i])    # ❌ IndexError if i goes out of range
for i in range(len(names)):
    print(f"{i+1}. {names[i]}")


# ========== Mistake 4: Using loop variable after loop ==========
numbers = [10, 20, 30, 40]

for num in numbers:
    doubled = num * 2

# print(num)         # ❌ May give last value, but not reliable
print("Mistake 4 fixed - Last doubled value:", doubled)


# ========== Mistake 5: Incorrect accumulation ==========
marks = [85, 90, 78]

# total = marks[0]   # Wrong initialization
total = 0
for mark in marks:
    total = total + mark          # Correct: total += mark

print("Mistake 5 fixed - Total:", total)


# ========== Mistake 6: Looping over dict instead of list of dicts ==========
students = [
    {"name": "Aarav", "cgpa": 8.5},
    {"name": "Priya", "cgpa": 9.1}
]

# for student in students[0]:     # ❌ Loops over keys of first dict only
for student in students:          # ✅ Correct
    print(student["name"], student["cgpa"])


# ========== Mistake 7: Not using enumerate() when index is needed ==========
subjects = ["Math", "Physics", "Chemistry"]

for i, subject in enumerate(subjects):
    print(f"{i+1}. {subject}")


print("\n=== All common for loop mistakes covered ===")
```

```python
# =====================================================
# MODULE 3 PART 1 - FOR LOOP CODE TRACING QUIZ
# Predict the output of each question before running
# =====================================================

print("========== QUIZ START ==========\n")

# Q1: Basic accumulation using for loop
marks = [85, 90, 78, 92]
total = 0
for mark in marks:
    total += mark
print("Q1 - Total:", total)
print()


# Q2: Using range()
print("Q2:")
for i in range(3):
    print(i)
print()


# Q3: enumerate() with list
subjects = ["Math", "Physics", "Chemistry"]
print("Q3:")
for index, subject in enumerate(subjects):
    print(index, subject)
print()


# Q4: Looping over list of dictionaries
students = [
    {"name": "Aarav", "cgpa": 8.5},
    {"name": "Priya", "cgpa": 9.1}
]
print("Q4:")
for s in students:
    print(s["name"], s["cgpa"])
print()


# Q5: Finding maximum value using loop
scores = [78, 92, 85, 88, 75]
highest = 0
for score in scores:
    if score > highest:
        highest = score
print("Q5 - Highest Score:", highest)
print()


# Q6: Counting with condition
marks = [45, 78, 32, 90, 55, 88]
pass_count = 0
for mark in marks:
    if mark >= 50:
        pass_count += 1
print("Q6 - Passed Students:", pass_count)
print()


# Q7: Computing average inside loop
data = [85, 90, 78, 92, 88]
total = 0
count = 0
for value in data:
    total += value
    count += 1
average = total / count
print("Q7 - Average:", round(average, 2))
print()


# Q8: Combined logic (Name + Average)
records = [
    {"name": "Rahul", "scores": [85, 90, 78]},
    {"name": "Sneha", "scores": [92, 88, 95]}
]
print("Q8:")
for student in records:
    avg = sum(student["scores"]) / len(student["scores"])
    print(student["name"], "-> Average:", round(avg, 2))
print()


print("========== QUIZ END ==========")
```
